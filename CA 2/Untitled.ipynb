```python
import tkinter as tk


    
    root = tk.Tk()
    def displayinfo(self, root):
        self.root = root
        self.root.title("PAU Cafetaria Order System")

        self.menu = {
          "RICE/PASTA"
            "jollof Rice":350,
            "Coconut Fried Rice":350,
            "Jollof Spaghetti":350,

            "PROTEINS"
                "Sweet Chill Chicken":1100,
                "Grilled Chicken Wings":400,
                "Fried Beef":400,
                "Fried Fish":500,
                "Boiled Egg":200,
                "Sauteed Sausages":200,

            "SIDE DISHES"
                "Savoury Beans":350,
                "Roasted Sweet Potatoes":300,
                "Fried Plantains":150,
                "Mixed Vegetable Salad":150,
                "Boiled Yam ":150,

            "SOUPS & SWALLOWS"
                "Eba":100,
                "Poundo Yam":100,
                "Semo":100,
                "Atama Soup":450,
                "Egusi Soup":480,

            "BEVERAGES"
                "Water":200,
                "Glass Drink (35cl)":150,
                "PET Drink (35cl)":300,
                "PET Drink (50cl)":350,
                "Glass/Canned Malt":500,
                "Fresh yo":600,
                "Pineapple Juice":350,
                "Mango Juice":350,
                "Zobo Drink":350,

            
                

            
            
        }

        self.order = {}

        self.create_widgets()

    def create_widgets(self):
        self.label_menu = tk.Label(self.root, text="Menu:")
        self.label_menu.pack()

        for item, price in self.menu.items():
            label_item = tk.Label(self.root, text=f"{item}: N{price}")
            label_item.pack()

        self.label_name = tk.Label(self.root, text="Enter your name:")
        self.label_name.pack()
        self.entry_name = tk.Entry(self.root)
        self.entry_name.pack()

        self.label_order = tk.Label(self.root, text="Enter your order:")
        self.label_order.pack()
        self.entry_order = tk.Entry(self.root)
        self.entry_order.pack()

        self.label_quantity = tk.Label(self.root, text="Enter quantity:")
        self.label_quantity.pack()
        self.entry_quantity = tk.Entry(self.root)
        self.entry_quantity.pack()

        self.submit_button = tk.Button(self.root, text="Submit Order", command=self.calculate_total)
        self.submit_button.pack()

    def calculate_total(self):
        name = self.entry_name.get()
        order_item = self.entry_order.get()
        quantity = int(self.entry_quantity.get())

        if order_item in self.menu:
            price = self.menu[order_item]
            total_price = price * quantity
            self.order[order_item] = total_price

            total_charges = sum(self.order.values())
            discount = self.calculate_discount(total_charges)

            total_with_discount = total_charges - discount

            self.display_total(name, total_with_discount)

    def calculate_discount(self, total_charges):
        if total_charges < 1000:
            return 0
        elif 1000 <= total_charges < 2500:
            return total_charges * 0.1
        elif 2500 <= total_charges < 5000:
            return total_charges * 0.15
        else:
            return total_charges * 0.25

    def display_total(self, name, total_with_discount):
        self.root.withdraw()  # Hide the main window

        # Create a new window for displaying total charges
        self.total_window = tk.Toplevel()
        self.total_window.title("Total Charges")

        label_summary = tk.Label(self.total_window, text=f"Hello {name}, your total charges are N{total_with_discount}")
        label_summary.pack()

        close_button = tk.Button(self.total_window, text="Close", command=self.close_windows)
        close_button.pack()

    def close_windows(self):
        self.total_window.destroy()
        self.root.destroy()




```
